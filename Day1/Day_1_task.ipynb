```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/student_performance.csv")
```

```python
print("="*60)
print("STUDENT PERFORMANCE EXPLORER REPORT")
print("="*60)

print("\n[1] DATASET OVERVIEW")
print("-"*40)
print("Total Students: ", df.shape[0])  #or len(df)
print("Total Columns: ", df.shape[1])   #or len(df.columns)
print("Department: ", len(df['department'].unique()))

print("\n[2] DEPARTMENT-WISE STUDENT COUNT")
print("-"*40)
print(df.groupby('department')['name'].count().sort_values(ascending=False))

print("\n[3] SCORE ANALYSIS")
print("-"*40)
print("Highest Math Score: ", df['math_score'].max())
print("Lowest Math Score: ", df['math_score'].min())
print("Average Math Score: ", df['math_score'].mean().round(2))
print(f"Average Programming Score: {df['programming_score'].mean():.2f}")
print(f"Average Attendance: {df['attendance_percentage'].mean():.2f}%")

df['total_score'] = df['math_score'] + df['english_score'] + df['programming_score'] + df['science_score']
print("\n[4] GENDER-WISE ANALYSIS")
print("-"*40)
print(f"Scores by gender:\n{df.groupby('gender')['total_score'].mean().round(3)}")

print("\n[5] TOP 5 PERFORMERS")
print("-"*40)
top5 = df.sort_values('total_score', ascending=False).head()
for index, row in top5.iterrows():    #iterroes retirns the df as (index, row)
  print(f" {index} \t| {row['name']:<20}| {row['department']:<10} \t| Scores: {row['total_score']}")

print("\n[6] STUDENTS WHO LOW ATTENDANCE (Below 75%)")
print("-"*40)
low_att = df[df['attendance_percentage'] < 75][['name', 'department', 'attendance_percentage']]
if len(low_att) > 0:
  print(low_att.to_string(index=False))
else:
  print("All students have attendance above 75%")

print("\n" + "="*60)
print("Report Complete")
print("="*60)
```
