```python
cs_students = df[df['department'] == 'Computer Science']

print(f"Number of Computer Science students: {len(cs_students)} ")
print("\n\tComputer Science Students: ")
print(cs_students[['name', 'math_score', 'programming_score', 'attendance_percentage']])
```

```python
above_90 = cs_students[cs_students['attendance_percentage'] > 90 ]
print(above_90[['name', 'department', 'attendance_percentage']])

print("\nNumber of students:", len(above_90))
```

```python
df['total_score'] = df['math_score'] + df['english_score'] + df['science_score'] + df['programming_score']
toppers = df[df['total_score']/4 > 90 ]
print(toppers[['name', 'total_score']])

print("\nNumber of students:", len(toppers))
```

```python
unique_dept = df['department'].unique()
dept_avg = df.groupby('department')['total_score'].mean().round(2)
# .sort_values(ascending=False)
print("Avg score, dept wise: \n", dept_avg)
```
