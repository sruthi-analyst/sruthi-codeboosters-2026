```python
raw_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/messy_sales_data.csv")

print(f'Raw data loaded: {raw_df.shape[0]} rows, {raw_df.shape[1]}columns')
print(f'Columns: {raw_df.columns.tolist()}')
print("\n First 5 rows: ")
print(raw_df.head().to_string(index=False))
```

```python
print('='*55)
print('DATA QUALITY DOAGNOSIS REPORT')
print('='*55)

#1. MISSING VALUES
print("\n[1] MISSING VALUES per column:")
print(raw_df.isna().sum())

#2. DUPLICATES
print("\n[2] DUPLICATES:", raw_df.duplicated().sum())
# duplicates assigns number 1 only to the second occrance of thta value, and not the first occurance- it only accounts for the duplicate rows

#3. DATA TYPES
print("\n[3] DATA TYPES: ")
print(raw_df.dtypes)

#4. Unique Values in text columns (spot inconsistencies)
print("\n[4] UNIQUE CATAGORIES: ", raw_df['category'].unique())
print("[4] Sample Customer Names", raw_df['customer_name'].dropna().unique()[:8])
print("[4] Sample order_date values: ", raw_df['order_date'].dropna().unique()[:8])
```

```python
print("[5] Missing values in quantity: ", raw_df['quantity'].isna().sum())
```

```python
print('='*55)
print('CREATE A WORKING COPY (ETL Best Practice)')
print('='*55)

df = raw_df.copy()

print(f"Working copy created: {df.shape}")
print('raw_df is untouched - we can always reset by running df = raw_df.copy()')
```

```python
print('='*55)
print('Fix #1: Handle Missing Values')
print('='*55)

print("Before handling nulls:\t", df.isnull().sum().sum(), "total missing values")

print("\nNo. of null values in Custome Name: ", df['customer_name'].isnull().sum())
df['customer_name'].fillna('Unknown Customer', inplace=True)
print("Null values of 'customer_name' columns handled successfully! ")
print("No. of null values in Custome Name: ", df['customer_name'].isnull().sum())


print("\nNo. of null values in quantity: ", df['quantity'].isnull().sum())
median_qty = df['quantity'].median()
df['quantity'].fillna(median_qty, inplace=True)
print("Null values of 'quantity' columns handled successfully! ")
print("No. of null values in quantity: ", df['quantity'].isnull().sum())


print("\nNo. of null values in category: ", df['category'].isnull().sum())
df['category'].fillna('uncategorised', inplace=True)
print("Null values of 'category' columns handled successfully! ")
print("No. of null values in category: ", df['category'].isnull().sum())


print("\nNo. of null values in product: ", df['product'].isnull().sum())
df['product'].fillna('Unknown Product', inplace=True)
print("Null values of 'product' columns handled successfully! ")
print("No. of null values in product: ", df['product'].isnull().sum())

print("\nAfter handling nulls:\t", df.isnull().sum().sum(), "total missing values")
```

```python
#Adding duplicate values to df
duplicate_row1 = df.iloc[3].copy()
duplicate_row2 = df.iloc[6].copy()
duplicate_row3 = df.iloc[18].copy()
df = pd.concat([df, pd.DataFrame([duplicate_row1])], ignore_index=True)
df = pd.concat([df, pd.DataFrame([duplicate_row2])], ignore_index=True)
df = pd.concat([df, pd.DataFrame([duplicate_row3])], ignore_index=True)

duplicate_rows = df[df.duplicated(keep=False)]
print(f"Found {len(duplicate_rows)} rows that are duplicates:\n")
display(duplicate_rows)

print(df.duplicated().sum(), "duplicate rows")
```

```python
print('='*55)
print('Fix #2: Handle Duplicate Values')
print('='*55)

print("Before handling duplicates:\t", df.duplicated().sum(), "duplicate rows")
df.drop_duplicates(inplace=True)
print("After handling duplicates:\t", df.duplicated().sum(), "duplicate rows")
```

```python
print('='*55)
print('Fix #3: Parsing')
print('='*55)

print('Sample dates before parsing: ')
print(df['order_date'].head(8).to_list())

df['order_date'] = pd.to_datetime(
    df['order_date'],
    dayfirst=False,
    errors='coerce'
)
nat_count = df['order_date'].isnull().sum()
print(f'\nUnparseable data (NaT): {nat_count}')

df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.strftime('%B')

print('\nSample dates after parsing: ')
print(df[['order_date', 'year', 'month', 'month_name']].head(8))
```
